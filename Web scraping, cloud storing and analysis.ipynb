```python
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup


my_url = 'https://www.imdb.com/list/ls084462066/'

```

```python
#opening up connection, grabbing the page


uClient = uReq(my_url)

page_html = uClient.read()

uClient.close()
```

```python
#html parser

page_soup = soup(page_html, "html.parser")


# grabs each product

fab = page_soup.findAll("div",{"class":"lister-item mode-detail"})
```

```python
#number of items

len(fab)
```

```python
filename = "products.csv"
f = open(filename, "w")

headers = "Film Title, IMDB Rating, Run Time, Genre, Short Narration, Directer\n"

f.write(headers)
```

```python
for fa in fab:

	try:
        
		title = fa.div.img["alt"]
		rating = fa.findAll("span", {"class":"ipl-rating-star__rating"})[0].text
		runtime = fa.findAll("span", {"class":"runtime"})[0].text
		genre = fa.findAll("span", {"class":"genre"})[0].text.strip()
		description = fa.findAll("p", {"class":""})[0].text.strip()
		directer = fa.findAll("p", {"class":"text-muted text-small"})[1].a.text


		print ( " Film Title : " + title)
		print (" IMDB Rating : " + rating)
		#print (" Run Time : " + runtime)
		print (" Genre : " + genre)
		print (" Short Narration : " + description)
		print (" Directer : " + directer)


		f.write(title.replace(",", " ") + "," + rating + "," + runtime + "," + genre.replace(",", "|")  + "," + description.replace(",", " ") + "," + directer + "\n")

	except IndexError:
		print ("_")


f.close()
```
